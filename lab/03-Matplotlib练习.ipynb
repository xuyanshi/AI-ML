```python
import matplotlib.pyplot as plt
cities=['Âåó‰∫¨','‰∏äÊµ∑','Â§©Ê¥•','ÈáçÂ∫Ü']
gdp=[12406.8,13908.57,9386.87,9143.64]
plt.barh(cities,gdp,color='blue')
plt.title('‰∏≠ÂõΩÂõõ‰∏™Áõ¥ËæñÂ∏Ç2017Âπ¥‰∏äÂçäÂπ¥ÁöÑGDP')
plt.xlabel('GDPÔºà‰∫øÂÖÉÔºâ')
plt.ylabel('Áõ¥ËæñÂ∏Ç')
plt.tight_layout()
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['font.family']='sans-serif'
plt.show()
```

### 2. ‰ΩøÁî®np.linspace()ÂáΩÊï∞ÁîüÊàêt‚àà[0, 2ùúã]„ÄÇÁªôÂÆöùë•=16*sin‚Å°(ùë°)^3Âíåùë¶=13*cos‚Å°(ùë°)‚àí5*cos‚Å°(2ùë°)‚àí2*cos‚Å°(3ùë°)‚àícos‚Å°(4ùë°)Ôºå‰ΩøÁî®Á∫¢Á∫øÁîªÂá∫x-yÂõæË°®ÔºåÂπ∂ÁªôÂõæÊ†áÊ∑ªÂä†‰∏Ä‰∏™È¢òÁõÆÔºöHeart

```python
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,2*np.pi,1000)
x=16*np.sin(t)**3
y=13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t)
plt.figure(figsize=(8,6))
plt.plot(x,y,color="red",linewidth=2)
plt.title('Heart')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
